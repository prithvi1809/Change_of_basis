```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath
from cmath import sqrt
from sympy.integrals.trigonometry import trigintegrate
from sympy.abc import x,t
from sympy import Symbol,sin,cos,tan,cot,sec,csc
import random
```

```python
#complex basis
Points=400          #points to be considered for construction of original curve
X=np.linspace(0,90,Points)
Y=[]
for t in X:
    Y.append(abs(10*math.cos(0.15*t)))
#print("X",X)
#print("Y",Y)
input_vec=np.vectorize(complex)(X,Y)
```

```python
#Change of Basis
def change_of_basis(input_vec):
    l=len(input_vec)
    vec1=[]
    for k in range(l):
        w=[]
        for u in range(l):
            m=1j
            d= cmath.exp((-m*2*(cmath.pi)*k*u)/l)
            w.append(d)
        basis_vec=np.hstack(w)


        abc = np.vdot(input_vec,basis_vec)

        vec1.append(abc)
    change_vec=np.hstack(vec1)
    return change_vec
```

```python
# calling Change of basis for input vector function
change_vec=change_of_basis(input_vec)
```

```python
#Cretion of partial vectors with random points.
choice=320           #points to be considered for reconstruction
a=[i for i in range(len(change_vec))]
index_random=np.random.choice(a,choice,replace=False) #random index Number array of size choice
index_random=sorted(index_random)
partial_vec=[]    #random chosen p points from change _vec array for reconsruction into complex plain.
for i in index_random:
    partial_vec.append(change_vec[i])
```

```python
#RECONSTRUCTION
def reconstruction(partial_vec):
    l=len(partial_vec)
    temp_A=[]

    for k in range(l):
        m=[]
        for u in range(l):
            p=1j
            h= cmath.exp((p*2*(cmath.pi)*k*u)/l)
            m.append(h)
        p=np.hstack(m)

        t = (np.vdot(partial_vec,p))*(1/l)
        temp_A.append(t)
    rec_vec=np.hstack(temp_A)
    rec_vec=rec_vec[1:]
    return rec_vec
```

```python
#Calling Reconstruction function
rec_vec=reconstruction(partial_vec)
```

```python
#Plotting function
#100%
fig,axs=plt.subplots(1,2)
fig.set_figheight(5)
fig.set_figwidth(10)
axs[0].plot(input_vec.real,input_vec.imag)
axs[1].plot(rec_vec.real,rec_vec.imag)


axs[0].set_xlabel("X ----->")
axs[0].set_ylabel("Y ----->")
axs[0].set_title("Original")
axs[1].set_title("Reconstructed(100%)")
plt.show()
plt.plot(change_vec.real,change_vec.imag)
plt.title("Signature");
```
